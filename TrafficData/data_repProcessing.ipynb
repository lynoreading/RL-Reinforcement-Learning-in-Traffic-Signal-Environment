```python
data_traffic_volume_path = 'D:\\Master_ML\\RL\\eco-light\\traffic-volume-counts-1.csv'
data_traffic_volume = pd.read_csv(data_traffic_volume_path, dtype=str)
```

```python
data_traffic_volume.columns = data_traffic_volume.columns.str.strip()
data_traffic_volume.columns
```

```python
ends_with_space = data_traffic_volume.applymap(lambda x: str(x).endswith(' '))
ends_with_space
```

```python
data_traffic_volume = data_traffic_volume.applymap(lambda x: x.rstrip() if isinstance(x, str) else x)
ends_with_space = data_traffic_volume.applymap(lambda x: str(x).endswith(' '))
ends_with_space
```

```python
for column in data_traffic_volume.columns[0:2]:
    data_traffic_volume[column] = pd.to_numeric(data_traffic_volume[column], errors='coerce').fillna(0).astype(int)
# data_traffic_volume.iloc[0:, 0:2] = data_traffic_volume.iloc[0:, 0:2].astype(int)
data_traffic_volume.iloc[0:, 2:7] = data_traffic_volume.iloc[0:, 2:7].astype(str)
for column in data_traffic_volume.columns[7:]:
    data_traffic_volume[column] = pd.to_numeric(data_traffic_volume[column], errors='coerce').fillna(0).astype(int)
data_traffic_volume.dtypes
```

```python
rows_with_nan = data_traffic_volume[data_traffic_volume.isna().any(axis=1)]
rows_with_nan
```

```python
data_traffic_volume['day_total'] = data_traffic_volume.iloc[:, 7:].sum(axis=1)
data_traffic_volume['day_mean'] = data_traffic_volume['day_total'] / 24
data_traffic_volume.columns
```
